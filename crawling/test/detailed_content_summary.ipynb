```python
import requests
from bs4 import BeautifulSoup

url = "https://www.jobkorea.co.kr/Recruit/GI_Read/47323347?rPageCode=SL&logpath=21&sn=6&sc=611"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36"
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
soup
```

```python
body_text = soup.body.get_text(separator='\n', strip=True)
body_text
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

model = ChatOpenAI(
    model = 'gpt-4.1-nano',
    temperature = 0.1,
)

prompt = PromptTemplate(
    template='''
    너는 채용공고문을 요약해주는 친절한 챗봇이야.
    다음 채용 공고문을 아래와 같이 3줄로 요약해줘.

    [핵심 요약]
    - 지역, 직무, 주요 조건만 포함한 한 문장
    - 예: "서울에서 마케팅 직원을 모집하며, 정규직, 연봉 3000만원"

    [자격요건]
    - 지원 자격 조건을 한 문장으로 요약
    - 예: "마케팅 관련 경력 2년 이상, 대졸 이상"

    [근무조건]
    - 근무시간, 형태, 복지 등을 한 문장으로 요약
    - 예: "주 5일 근무, 정규직, 재택 가능, 식대 지원"

    공고문:
    {content}
    ''',
    input_variables=['content']
)

output_parser = StrOutputParser()

chain = prompt | model | output_parser
result = chain.invoke({'content': body_text})
```

```python
result
```

```python
def get_text_or_none(tag):
    return tag.get_text(strip=True) if tag else None

def get_all_texts(tags):
    return [t.get_text(strip=True) for t in tags] if tags else None

def get_next_dd_by_dt(text):
    dt = soup.select_one(f'dt:-soup-contains("{text}")')
    return dt.find_next_sibling('dd') if dt else None

result = {}

# 회사이름
company_name = soup.select_one('span.coName')
result['회사이름'] = get_text_or_none(company_name)

# 경력
career = get_next_dd_by_dt('경력')
result['경력'] = get_text_or_none(career)

# 학력
education = get_next_dd_by_dt('학력')
result['학력'] = get_text_or_none(education)

# 스킬
skill = get_next_dd_by_dt('스킬')
if skill:
    skills = [s.strip() for s in skill.get_text(strip=True).split(',')]
else:
    skills = None
result['스킬'] = skills

# 우대
preference_spans = soup.select('dl#dlPref dd span.pref')
result['우대'] = get_all_texts(preference_spans)

# 고용형태
employment_list = []
employment_lis = soup.select('dt:-soup-contains("고용형태") + dd li')
for li in employment_lis:
    employment_list.append(li.get_text(strip=True))
result['고용형태'] = employment_list if employment_list else None

# 급여
salary = get_next_dd_by_dt('급여')
result['급여'] = get_text_or_none(salary)

# 지역
location = soup.select_one('dt:-soup-contains("지역") + dd a')
result['지역'] = get_text_or_none(location)

# 시간
working_time = get_next_dd_by_dt('시간')
result['시간'] = get_text_or_none(working_time)

# 직급
position = get_next_dd_by_dt('직급')
result['직급'] = get_text_or_none(position)

# 직책
role = get_next_dd_by_dt('직책')
result['직책'] = get_text_or_none(role)
```
